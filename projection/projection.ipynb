```python
import math

import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
```

## 1. Range and Height

### a. Parameterized by velocity and theta

```python
max_theta = 90
max_v = 30
g = 9.81
```

```python
def compute_H(velocity, theta):
    return (velocity * np.sin(np.radians(theta)))**2 / 2 * g


def compute_R(velocity, theta):
    return (velocity**2 * np.sin(np.radians(2 * theta))) / g
```

```python
# Obtain meshgrid points
vel = np.linspace(0, 6, 30)
theta = np.linspace(0, 6, 30)
vel, theta = np.meshgrid(vel, theta)
```

```python
# Plot 3D surface plot
fig = plt.figure(figsize=(16, 18))

# For height
height = compute_H(vel, theta)
ax1 = fig.add_subplot(1, 2, 1, projection="3d")
ax1.set_xlabel("velocity")
ax1.set_ylabel("theta")
ax1.set_title("Height")
a1 = ax1.plot_surface(vel, theta, height, 
                      label=r"$\frac{ ( V \sin\theta)^2}{2g}$")
a1._edgecolors2d = a1._edgecolors3d
a1._facecolors2d = a1._facecolors3d
ax1.legend(prop={"size": 16})

# For range
R = compute_R(vel, theta)
ax2 = fig.add_subplot(1, 2, 2, projection="3d")
ax2.set_xlabel("velocity")
ax2.set_ylabel("theta")
ax2.set_title("Range")
a2 = ax2.plot_surface(vel, theta, R, 
                     label=r"$\frac{ V^2 \sin2\theta }{g}$")
a2._edgecolors2d = a2._edgecolors3d
a2._facecolors2d = a2._facecolors3d
ax2.legend(prop={"size": 16})

plt.savefig("plot.pdf")
```

### b. Line plot

```python
plt.style.use("ggplot")

plt.figure(figsize=(8, 6))
degree = list(range(10, 90, 10))
for idx in range(len(degree)):
    plt.plot([0, 20 * 4 / math.tan(math.radians(degree[idx]))], 
             [0, 20], label=f"{degree[idx]}$^\circ$")
    
plt.xlabel("R")
plt.ylabel("H")
plt.title(r"$R=4H \cot \theta$")
plt.grid(False)
plt.legend()

plt.savefig("line.pdf")
```
